```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.amazon.in/s?k=keyboard&page=2&crid=1UY6W0VLQ60A7&qid=1706706625&sprefix=keyboard%2Caps%2C256&ref=sr_pg_1',headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

```python
links=soup.find_all('span',class_="a-size-medium a-color-base a-text-normal")
len(links)
```

```python
prod_na=[]
for tag in links:
    prod_na.append(tag.text.strip())
```

```python
prices=soup.find_all('span',class_="a-price-whole")
price=[]
for cost in prices:
    price.append(cost.text.strip())
```

```python
dates=soup.find_all('span', class_="a-color-base a-text-bold")
all_dates=[]
for dt in dates:
    all_dates.append(dt.text.strip())
```

```python
stars=soup.find_all('i',class_="a-icon a-icon-star-small a-star-small-4 aok-align-bottom")
rating=[]
for st in stars:
    rating.append(st.text.strip())
```

```python
df1 = pd.DataFrame({'Rating': rating})
df1
```

```python
df2 = pd.DataFrame({'Price': price})
df2.reset_index(inplace=True)
df1.reset_index(inplace=True)
df2
```

```python
merged_df = pd.merge(df1, df2)
merged_df = merged_df.reset_index(drop=True)
```

```python
DF1=merged_df
```

```python
df3=pd.DataFrame({'Name': prod_na})
df3.reset_index(inplace=True)
df3
```

```python
merged_df = pd.merge(DF1, df3)
```
